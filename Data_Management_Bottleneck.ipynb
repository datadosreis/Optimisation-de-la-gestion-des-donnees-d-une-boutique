#### Mesures de dispersion

Le boxplot permet de représenter schématiquement une distribution, en incluant sa dispersion. La boîte est délimitée par Q1 et Q3 , la médiane est représentée à l’intérieur de la boîte. On dessine ensuite des moustaches à cette boîte, qui vont de la valeur minimale à la valeur maximale... à condition que la moustache (d'un côté ou de l'autre) ne mesure pas plus de 1,5 fois l'écart inter-quartiles. Si certaines valeurs sont au-dessous de  Q1−1.5IQ ou au-dessus de Q3+1.5IQ , alors on les considère comme des outliers.

Le boxplot ci-dessous nous permet d'ores et déjà d'identifier que nous avons un certain nombre d'outliers. 

```python
#boxplot
plt.figure(figsize=(16,5))
CA_par_produit.boxplot(column="price", vert=False)
plt.show()
```

#### Mesures de forme

```python
print("Skewness empirique =",round(CA_par_produit['price'].skew(),2),)
```

Le skewness est une mesure d'asymétrie. L’asymétrie d’une distribution traduit la régularité (ou non) avec laquelle les observations se répartissent autour de la valeur centrale. On interprète cette mesure de cette manière :

- Si γ1=0 alors la distribution est symétrique.

- Si γ1>0 alors la distribution est étalée à droite.

- Si γ1<0 alors la distribution est étalée à gauche.

Dans notre cas, le skewness nous montre que les valeurs se répartissent principalement à droite (vers les prix les plus élevés) et donc que **la distribution n'est pas symétrique.** La mesure de l'aplatissement via le kurtosis empirique n'est donc pas pertinente.

### Distribution empirique des prix

On souhaite regrouper les valeurs en classes.
On utilise donc la méthode  hist(), qui s'occupe elle-même de regrouper les valeurs en classes.
Enfin, on spécifie le nombre de classes voulues grâce au mot clé  bins  : ici 20.

```python
plt.figure(figsize=(16,5))
plt.hist(CA_par_produit["price"],ec="black",color='blue',bins=20)
plt.title("Distribution empirique des prix",fontsize=12)
plt.xlabel("Prix", fontsize=10)
plt.ylabel("Count", fontsize=10)
plt.show()
```

La distribution empirique de la variable "prix" est clairement asymétrique.
Comme nous avons pu le voir précédemment en calculant le skewness, la distribution est clairement étalée à droite avec quelques prix élevés pour certains produits.

Il est donc important de s'intéresser à ces outliers afin de voir si nous devons les considérer comme des valeurs aberrantes ou non.

### Détection des outliers

#### Méthode de la distance interquartile

```python
Q1 = CA_par_produit["price"].quantile(.25)
Q3 = CA_par_produit["price"].quantile(.75)
Ecart_interquartile = Q3 - Q1

Outliers_Q1 = round(Q1 - 1.5 * Ecart_interquartile,2)
print("La limite inférieure est de",Outliers_Q1,"€")
#Calcul de la limite à partir de laquelle on considère une valeur comme outliers_min
Outliers_Q3 = round(Q3 + 1.5 * Ecart_interquartile,2)
print("La limite supérieure est de",Outliers_Q3,"€")
#Calcul de la limite à partir de laquelle on considère une valeur comme outliers_max
```

```python
df_Outliers = CA_par_produit.loc[(CA_par_produit["price"] > Outliers_Q3) | (CA_par_produit["price"] < Outliers_Q1) ]
df_Outliers.sort_values("CA_Produit", ascending=False)
```

```python
print("Nous identifions ici",df_Outliers.shape[0],"outliers.")
```

```python
n,m,patches = plt.hist(CA_par_produit["price"],ec="black",bins=20)

for i in range (7,20):
    patches[i].set_fc("red")

plt.title("Distribution empirique des prix",fontsize=12)
plt.xlabel("Prix en €", fontsize=10)
plt.ylabel("Count", fontsize=10)
plt.axvline(Outliers_Q3, color="black", linestyle="--")
plt.legend({"Limite Outliers":Outliers_Q3})
plt.show()
```

#### Méthode du Z-Score

Rappel sur la méthode du Z-Score:

**z = (x – μ)/σ** avec μ = moyenne de la population et σ = écart-type de la population

Une distribution normale est représentée ci-dessous et on estime que:

- 68% des points de données se situent entre +/- 1 écart-type.
- 95 % des points de données se situent entre +/- 2 écarts types.
- 99,7 % des points de données se situent entre +/- 3 écarts types.

Pour plus de sensibilité, nous utiliserons un seuil de 2 pour considérer un produit comme outlier.

```python
df_zscore = CA_par_produit.copy()
#Copie du df utilisé précedemment dans un nouveau df dédié au z_score
threshold=2
#Définition du seuil à partir duquel on considére les valeurs comme outliers
df_zscore['z_score'] = stats.zscore(df_zscore['price'])
#Création de la colonne z_score à partir de la librairie scipy.stats qui permet de calculer le z-score
df_zscore.loc[df_zscore['z_score'] > threshold ].sort_values("CA_Produit", ascending=False)
#Filtre uniquement sur les produits dépassant le seuil de 2 au niveau du z-score et tri décroissant par CA
```

```python
print("En utilisant la méthode du Z-Score, nous identifions",df_zscore.loc[df_zscore['z_score'] > 2 ].shape[0],"outliers.")
```

```python
plt.figure(figsize=(16,5))
sns.histplot(data=df_zscore['z_score'],bins=30, kde=True).set(xlim=[-3,8])
plt.axvline(threshold, color="red", linestyle="--", label='Limite Outliers')
plt.title("Analyse des Outliers via Z-Score", fontsize=12)
plt.legend()
plt.show()
```
